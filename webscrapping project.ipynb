```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as s
from urllib.request import urlopen as u
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
link="https://www.flipkart.com/search?q=dell+laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY"
pagecontent=u(link)
pagecode=pagecontent.read()
```

```python
page_file=s(pagecode)
print(page_file.prettify())
```

```python
modelname=page_file.find("div",{"class":'_4rR01T'})
print(modelname.text)
```

```python
productname=[]
modelname=page_file.findAll("div",{"class":'_4rR01T'})
for i in modelname:
    p=i.text
    productname.append(p)
```

```python
productname
```

```python
rating=[]
ratingname=page_file.findAll("div",{"class":'_3LWZlK'})
for i in ratingname:
    p=i.text
    rating.append(p)
```

```python
rating
```

```python
len(rating)
```

```python
del rating[:14]
```

```python
len(rating)
```

```python
offers=[]
of=page_file.findAll("div",{"class":'_3Ay6Sb'})
for i in of:
    p=i.text
    offers.append(p)
```

```python
offers
```

```python
len(offers)
```

```python
del offers[:2]
len(offers)
```

```python
price=[]
pr=page_file.findAll("div",{"class":'_30jeq3 _1_WHN1'})
for i in pr:
    p=i.text
    price.append(p)
```

```python
d=({'productname':productname,'rating':rating,'price':price,'offers':offers})
df=pd.DataFrame(d)
```
